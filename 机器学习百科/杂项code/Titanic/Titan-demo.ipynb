```python
import pandas as pd

train = pd.read_csv('../input/titanic/train.csv')
test = pd.read_csv('../input/titanic/test.csv')
```

```python
print(train.info)
```

```python
print(test.info)
```

```python
print(train.info())
print(test.info())
```

```python
selected_features = ['Pclass', 'Sex', 'Age', 'Embarked', 'SibSp', 'Parch', 'Fare']
X_train = train[selected_features]
X_test = test[selected_features]
y_train = train['Survived']

print(X_train['Embarked'].value_counts())
print(X_test['Embarked'].value_counts())
```

```python
X_train.shape, X_test.shape
```

```python
print(644 + 168 + 77, 270 + 102 + 46)
```

```python
X_train['Embarked'].fillna('S', inplace=True)
# X_test['Embarked'].fillna('S', inplace=True)

X_train['Age'].fillna(X_train['Age'].mean(), inplace=True)
X_test['Age'].fillna(X_test['Age'].mean(), inplace=True)
X_test['Fare'].fillna(X_test['Fare'].mean(), inplace=True)

X_train.info()
```

```python
X_test.info()
```

```python
from sklearn.feature_extraction import DictVectorizer
dict_vec = DictVectorizer(sparse=False)
X_train = dict_vec.fit_transform(X_train.to_dict(orient='record'))
dict_vec.feature_names_
```

```python
X_test = dict_vec.transform(X_test.to_dict(orient='record'))
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier()
xgbc = XGBClassifier()

cross_val_score(rfc, X_train, y_train, cv=5).mean()
```

```python
cross_val_score(xgbc, X_train, y_train, cv=5).mean()
```

```python
for dirname, _, filenames in os.walk('/kaggle/output'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
rfc.fit(X_train, y_train)
rfc_y_predict = rfc.predict(X_test)
rfc_submission = pd.DataFrame({'PassengerId': test['PassengerId'],
                              'Survived': rfc_y_predict})
rfc_submission.to_csv('./rfc_submission.csv', index=False)
```

```python
xgbc.fit(X_train, y_train)
xgbc_y_predict = xgbc.predict(X_test)
xgbc_submission = pd.DataFrame({'PassengerId': test['PassengerId'],
                               'Survived': xgbc_y_predict})
xgbc_submission.to_csv('./xgbc_submission.csv', index=False)
```

```python
from sklearn.model_selection import GridSearchCV
params = {'max_depth': range(2, 7),
         'n_estimators': range(100, 1100, 200),
         'learning_rate': [0.05, 0.1, 0.25, 0.5, 1.0]}

xgbc_best = XGBClassifier()
gs = GridSearchCV(xgbc_best, params, n_jobs=-1, cv=5, verbose=1)
gs.fit(X_train, y_train)
```

```python
print(gs.best_score_)
print(gs.best_params_)
```

```python
xgbc_best_y_predict = gs.predict(X_test)
xgbc_best_submission = pd.DataFrame({'PassengerId': test['PassengerId'],
                                    'Survived': xgbc_best_y_predict})
xgbc_best_submission.to_csv('./xgbc_best_submission.csv', index=False)
```
